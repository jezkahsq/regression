## 模型一：线性回归
### <font color ='SteelBlue'>给定数据集</font>
\begin{equation}D=\left\{\left(\boldsymbol{x}_{1}, y_{1}\right),\left(\boldsymbol{x}_{2}, y_{2}\right), \ldots,\left(\boldsymbol{x}_{m}, y_{m}\right)\right\}\end{equation}
### <font color ='SteelBlue'>样本值</font>\begin{equation}\boldsymbol{x}_{i}=\left(x_{i 1};\right.\left.x_{i 2}; \ldots ; x_{i d}\right), y_{i} \in \mathbb{R}\end{equation}

### <font color ='SteelBlue'>目的</font>
试图学得一个线性模型以尽可能准确地预测，即试图学得\begin{equation}f(\boldsymbol{x})=w_{1} x_{1}+w_{2} x_{2}+\ldots+w_{d}x_{d}+b\end{equation}使得\begin{equation}f\left(x_{i}\right) \simeq y_{i}\end{equation}

### <font color ='SteelBlue'>要学习的参数</font>
\begin{equation}\boldsymbol{w}=\left(w_{1} ; w_{2} ; \ldots ; w_{d}\right)\text{和}b\end{equation}
<font color ='SteelBlue'>很好的可解释性</font>：参数直观的表达了各个变量在预测中的重要性

### 如何衡量f(x)与y之间的差别？——最小化均方误差
**<font color ='SteelBlue'>均方误差</font>**是回归任务中最常用的性能度量
\begin{equation}\left(w^{*}, b^{*}\right)=\underset{(w, b)}{\arg \min } \sum_{i=1}^{m}\left(f\left(x_{i}\right)-y_{i}\right)^{2}\end{equation}

## 模型二：岭回归
\begin{equation}J(\theta)=\frac{1}{2 m} \sum_{i=1}^{m}\left[\left(\left(f\left(x_{i}\right)-y_{i}\right)^{2}+\lambda \sum_{j=1}^{d} \ w_{j}^{2}\right)\right]\end{equation}
### 岭回归通过对系数的大小施加惩罚项来解决普通最小二乘法的一些问题：

 + 当特征数大于样本数时，在模型推导过程中会出现 $\mathbf{X}^{T} \mathbf{X}$ 不可逆，故不能直接用线性回归(特征高度相关可使用岭回归)
 + 引入惩罚项，防止过拟合

## 模型三：贝叶斯岭回归
贝叶斯学习过程：在前一个训练集合的后验概率上，乘以新的测试样本点的似然估计，得到新的集合的后验概率
\begin{equation}p\left(\theta \mid D^{n}\right) \propto p\left(x_{n} \mid \theta\right) p\left(\theta \mid D^{n-1}\right)\end{equation}

## 模型四：支持向量回归
### <font color ='SteelBlue'>模型</font>：
\begin{equation}f(\boldsymbol{x})=w_{1} x_{1}+w_{2} x_{2}+\ldots+w_{d}x_{d}+b\end{equation}
红色显示出 $\epsilon$ 间隔带，落入其中的样本不计算损失(即落入间隔带的训练样本，被认为是预测正确的)

### <font color ='SteelBlue'>优化目标</font>：
\begin{equation}\min _{w, b, x, y} \frac{1}{2}\|w\|^{2}+C \sum_{i=1}^{N} l_{\epsilon}\left(w, b, x_{i}, y_{i}\right)\end{equation}
其中，\begin{equation}l_{\epsilon}\left(w, b, x_{i}, y_{i}\right)=\left\{\begin{array}{cc}
0, & \left|f\left(x_{i}\right)-y_{i}\right| \leq \epsilon \\
\left|f\left(x_{i}\right)-y_{i}\right|-\epsilon, & \left|f\left(x_{i}\right)-y_{i}\right|>\epsilon
\end{array}\right.\end{equation}

## 回归模型评估指标
- **解释方差**（Explianed variance score）\begin{equation}\text {Explianed_variance}(y, \hat{y})=1-\frac{\operatorname{Var}\{y-\hat{y}\}}{\operatorname{Var}\{y\}}\end{equation}
- **绝对平均误差**（Mean absolute error）\begin{equation}M A E(y, \hat{y})=\frac{1}{n} \sum_{i=0}^{n-1}\left|y_{i}-\hat{y}\right|\end{equation}
- **均方误差**（Mean squared error）\begin{equation}M S E(y, \hat{y})=\frac{1}{n} \sum_{i=0}^{n-1}\left(y_{i}-\hat{y}\right)^{2}\end{equation}
- **决定系数**（$R^{2}$score）\begin{equation}R^{2}(y, \hat{y})=1-\frac{\sum_{i=0}^{n-1}\left(y_{i}-\hat{y}_{i}\right)^{2}}{\sum_{i=0}^{n-1}\left(y_{i}-\bar{y}\right)^{2}}\end{equation}
